```python
def func1(text):
    no_punc = [char for char in text if char not in string.punctuation]
    no_punc = ''.join(no_punc)
    return no_punc
```

```python
data['text'] = data['text'].apply(func1)
```

```python
data['text'] = data['text'].str.lower()
```

```python
def func2(text):
    stop_words = [word for word in text.split() if word.lower() not in stopwords.words('english')]
#     stop_words = ''.join(stop_words)
    return stop_words
```

```python
# data['text'].apply(func2)
```

```python
data['text'] = data['text'].apply(func2)
```

```python
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def func3(text):
    stem_words = [word for word in text if ps.stem(word)]
#     stop_words = ''.join(stop_words)
    return stem_words
```

```python
data['text'] = data['text'].apply(func3)
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lem = WordNetLemmatizer()
```

```python
def func4(text):
    lem_words = [word for word in text if lem.lemmatize(word)]
#     stop_words = ''.join(stop_words)
    return lem_words
```

```python
data['text'] = data['text'].apply(func4)
```

```python
data.head()
```

```python
def func5(text):
    str_words = [word for word in text]
    str_words = ' '.join(str_words)
    return str_words
```
